```python
import tkinter as tk
from tkinter import ttk, messagebox
from math import sin, cos, tan, sqrt, log, log10, exp, factorial

def calculate_basic():
    try:
        num1 = float(entry1.get())
        num2 = float(entry2.get())
    except ValueError:
        messagebox.showerror("Error", "Please enter valid numbers.")
        return

    operation = combo.get()
    if operation == '+':
        result = num1 + num2
    elif operation == '-':
        result = num1 - num2
    elif operation == '*':
        result = num1 * num2
    elif operation == '/':
        result = num1 / num2
    elif operation == '%':
        result = (num2 / num1) * 100

    result_label.config(text="Result: " + str(result))

def clear_basic():
    entry1.delete(0, tk.END)
    entry2.delete(0, tk.END)
    result_label.config(text="Result: ")

def show_definition_basic():
    selected_operation = combo.get()
    definitions = {
        '+': "Addition: Adds two numbers.",
        '-': "Subtraction: Subtracts one number from another.",
        '*': "Multiplication: Multiplies two numbers.",
        '/': "Division: Divides one number by another.",
        '%': "Percentage: Calculates the percentage of the second number in relation to the first number.",
    }
    definition = definitions.get(selected_operation, "No definition found.")
    messagebox.showinfo("Operation Definition", definition)

def calculate_scientific():
    try:
        num = float(entry3.get())
    except ValueError:
        messagebox.showerror("Error", "Please enter a valid number.")
        return

    operation = combo2.get()
    if operation == 'sin':
        result = sin(num)
    elif operation == 'cos':
        result = cos(num)
    elif operation == 'tan':
        result = tan(num)
    elif operation == 'sqrt':
        result = sqrt(num)
    elif operation == 'log':
        result = log(num)
    elif operation == 'log10':
        result = log10(num)
    elif operation == 'exp':
        result = exp(num)
    elif operation == 'factorial':
        result = factorial(num)

    result_label2.config(text="Result: " + str(result))

def clear_scientific():
    entry3.delete(0, tk.END)
    result_label2.config(text="Result: ")

def show_definition_scientific():
    selected_operation = combo2.get()
    definitions = {
        'sin': "Sine: Calculates the sine of the input angle (in radians).",
        'cos': "Cosine: Calculates the cosine of the input angle (in radians).",
        'tan': "Tangent: Calculates the tangent of the input angle (in radians).",
        'sqrt': "Square Root: Calculates the square root of the input number.",
        'log': "Logarithm: Calculates the natural logarithm (base e) of the input number.",
        'log10': "Logarithm (Base 10): Calculates the logarithm (base 10) of the input number.",
        'exp': "Exponential: Calculates the exponential of the input number.",
        'factorial': "Factorial: Calculates the factorial of the input number.",
    }
    definition = definitions.get(selected_operation, "No definition found.")
    messagebox.showinfo("Operation Definition", definition)

window = tk.Tk()
window.title("Calculator")

frame = ttk.Frame(window, padding=20)
frame.grid()

basic_label = ttk.Label(frame, text="Basic Calculator", font=('Arial', 16, 'bold'))
basic_label.grid(row=0, columnspan=4, pady=10)

entry1 = ttk.Entry(frame, width=10)
entry1.grid(row=1, column=0, padx=5, pady=5)

combo = ttk.Combobox(frame, values=['+', '-', '*', '/', '%'], width=8)
combo.current(0)
combo.grid(row=1, column=1, padx=5, pady=5)

entry2 = ttk.Entry(frame, width=10)
entry2.grid(row=1, column=2, padx=5, pady=5)

calculate_button = ttk.Button(frame, text="Calculate", command=calculate_basic)
calculate_button.grid(row=1, column=3, padx=5, pady=5)

clear_button = ttk.Button(frame, text="Clear", command=clear_basic)
clear_button.grid(row=2, column=0, padx=5, pady=5)

definition_button = ttk.Button(frame, text="Definition", command=show_definition_basic)
definition_button.grid(row=2, column=1, padx=5, pady=5)

result_label = ttk.Label(frame, text="Result: ", font=('Arial', 14))
result_label.grid(row=3, columnspan=4, padx=5, pady=10)

frame2 = ttk.Frame(window, padding=20)
frame2.grid()

scientific_label = ttk.Label(frame2, text="Scientific Calculator", font=('Arial', 16, 'bold'))
scientific_label.grid(row=0, columnspan=4, pady=10)

entry3 = ttk.Entry(frame2, width=10)
entry3.grid(row=1, column=0, padx=5, pady=5)

combo2 = ttk.Combobox(frame2, values=['sin', 'cos', 'tan', 'sqrt', 'log', 'log10', 'exp', 'factorial'], width=8)
combo2.current(0)
combo2.grid(row=1, column=1, padx=5, pady=5)

calculate_button2 = ttk.Button(frame2, text="Calculate", command=calculate_scientific)
calculate_button2.grid(row=1, column=2, padx=5, pady=5)

clear_button2 = ttk.Button(frame2, text="Clear", command=clear_scientific)
clear_button2.grid(row=2, column=0, padx=5, pady=5)

definition_button2 = ttk.Button(frame2, text="Definition", command=show_definition_scientific)
definition_button2.grid(row=2, column=1, padx=5, pady=5)

result_label2 = ttk.Label(frame2, text="Result: ", font=('Arial', 14))
result_label2.grid(row=3, columnspan=4, padx=5, pady=10)

window.mainloop()
```
